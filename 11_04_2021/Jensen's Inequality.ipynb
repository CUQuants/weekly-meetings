```python
import numpy as np
import matplotlib.pyplot as plt
```

Log-normal stock price process:

```python
vol = 0.01
mu = 0.001
n = 100
T = 100
chg = np.random.normal(mu, vol, (T, n))
chg_cum = np.concatenate((np.zeros((1, n)), np.cumsum(chg, 0)))
price = 100*np.exp(chg_cum)
```

```python
plt.plot(price, linewidth = 0.5)
plt.plot(np.average(price, axis = 1), color = 'black', linewidth = 2)
plt.xlabel('t')
plt.ylabel('Price S')
plt.show()
```

```python
c_fn = lambda s, k: np.maximum(s-k, 0)
```

```python
strike = 105
plt.plot(np.average(c_fn(price, strike), axis=1), label=r'$E[(S-{})^+]$'.format(strike))
plt.plot(c_fn(np.average(price, axis = 1), strike), label=r'$(E[S]-{})^+$'.format(strike))
plt.legend()
plt.xlabel('t')
plt.ylabel('Call Payoff')
plt.show()
```

Note that at each point in time, the expected payoff of the call is always greater than or equal to the payoff of the call at the expected value of the stock.
