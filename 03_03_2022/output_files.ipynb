```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Overall Goal
I recently was reading the Bridgewater Associates report about Cryptocurrencies becoming more mainstraim and its future role in institutional portfolios. Something that Bridgewater had highlighted was the possibility of hedge funds using long short methods. This was something that I explored but I came across a major problem, which is that long / short portfolios are built upon determining risk premias for equities. Given that risk premias are already hard to understand in general the risk premias that occur in cryptocurrency are hard to find. 

I also wanted to convey that the sector (I guess its more of an asset class) doesn't have a high amount of correlation, which means that the longs don't necessarily offset the shorts (or vice versa) which is a necessity for long / short funds to be market neutral. I tried using technology stocks as an example but to my dismay, it was also not correlated. In this case I am using stock returns of daily and weekly returns and measuring correlation via correlation and covariance matrix. In this case I review cryptocurrency, technology stocks, and utility stocks. I'm hoping that utility stocks returns are relatively the same. 

```python
crypto_daily_prices = pd.read_csv("crypto_daily_prices.csv", index_col = 0)
crypto_weekly_prices = pd.read_csv("crypto_weekly_prices.csv", index_col = 0)
```

```python
def make_plots(df, security_type, frequency):

    fig, axes = plt.subplots(1, 2, figsize = (30, 10))
    df = df.pct_change().dropna()

    df_corr = df.corr()
    df_cov = df.cov()

    sns.heatmap(df_corr, ax = axes[0])
    axes[0].set_title("{} {} correlation matrix".format(security_type, frequency))

    sns.heatmap(df_cov, ax = axes[1])
    axes[1].set_title("{} {} covariance matrix".format(security_type, frequency))

    plt.show()
```

```python
make_plots(crypto_daily_prices, "crypto", "daily")
```

```python
make_plots(crypto_weekly_prices, "crypto", "weekly")
```

```python
tech_daily_prices = pd.read_csv("tech_daily_prices.csv", index_col = 0)
tech_weekly_prices = pd.read_csv("tech_weekly_prices.csv", index_col = 0)
```

```python
make_plots(tech_daily_prices, "tech", "daily")
```

```python
make_plots(tech_weekly_prices, "tech", "weekly")
```
