```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
#download the data
mnist = tf.keras.datasets.mnist

#load the data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
#sample output
plt.figure(figsize = (14,14))
plt.imshow(x_train[0])
```

```python
#now normalize the data
x_train = tf.keras.utils.normalize(x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_train, axis = 1)
```

```python
#now make the neural network

#model type
model = tf.keras.models.Sequential()

#the first layer (input) is a flatten layer
model.add(tf.keras.layers.Flatten())

#then we'll have 2 hidden layers with 128 neurons and use rectified linear as our activation function
model.add(tf.keras.layers.Dense(128, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(128, activation = tf.nn.relu))

#then make the output layer which requires a softmax activation function
model.add(tf.keras.layers.Dense(10, activation = tf.nn.softmax))
```

```python
#the compile the model 
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])

#now fit the model
model.fit(x_train, y_train, epochs = 10)
```

```python
#make our predictions
predictions = model.predict([x_test])
```

```python
#find 10 random numbers 
rand_nums = np.random.randint(len(predictions), size = 10)

for i in rand_nums:
    
    plt.figure(figsize = (10,10))
    plt.title(np.argmax(predictions[i]))
    plt.imshow(x_test[i])
```
