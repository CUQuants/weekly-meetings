```python
import numpy as np
from itertools import product

import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
#funciton that makes the porabola
def func(x):
    return x**2 - 2*x - 3
```

```python
def fprime(x):
    return 2*x - 2
```

```python
def plotFunc(x0):
    
    x = np.linspace(-5, 7, 100)
    
    plt.plot(x, func(x))
    plt.plot(x0, func(x0), 'ro')
    plt.xlabel("$x$")
    plt.ylabel("$f(x)$")
    plt.title("Objective Function")
```

```python
def plotPath(xs, ys, x0):
    
    plotFunc(x0)
    plt.plot(xs, ys, linestyle = '--', marker = 'o', color = 'orange')
    plt.plot(xs[-1], ys[-1], 'ro')
```

```python
x0 = -4
plotFunc(x0)
```

```python
def GradientDescentSimple(func, fprime, x0, alpha, tol = 1e-5, max_iter = 1000):
    
    xk = x0
    fk = func(xk)
    pk = -fprime(xk)
    
    #set everything to 0 and put it into the a list
    num_iter = 0
    curve_x = [xk]
    curve_y = [fk]
    
    #making sure that we haven't passed the max iter list and the tol check (not sure what that check is)
    while abs(pk) > tol and num_iter < max_iter:
        
        xk = xk  + alpha * pk
        fk = func(xk)
        pk = -fprime(xk)
        
        num_iter += 1
        curve_x.append(xk)
        curve_y.append(fk)
        
    if num_iter == max_iter:
        print('Gradient Descent does not converge')
        
    else: 
        print("Solution found:\n y = {:.4f}\n x={:.4f}".format(fk, xk))
        
        
    return curve_x, curve_y
```

```python
xs, ys = GradientDescentSimple(func, fprime, x0, alpha = 0.1)
plotPath(xs, ys, x0)
```

```python
xs, ys = GradientDescentSimple(func, fprime, x0, alpha = 0.9)
plotPath(xs, ys, x0)
```

```python
xs, ys = GradientDescentSimple(func, fprime, x0, alpha = 1e-4)
plotPath(xs, ys, x0)
```

```python
xs, ys = GradientDescentSimple(func, fprime, x0, alpha = 1.01, max_iter = 8)
plotPath(xs, ys, x0)
```
