```python
import pandas as pd
import matplotlib.pyplot as plt

from stochastic.processes.diffusion import OrnsteinUhlenbeckProcess
```

```python
plt.figure(figsize = (24, 10))
for i in range(10):
    ou_process = pd.Series(OrnsteinUhlenbeckProcess(0, 1).sample(1000, 1))
    ou_process.plot()
    
plt.grid(True)
plt.title("1000 Ornstein Uhlenbeck Process 0 speed reversion")
plt.show()
```

```python
plt.figure(figsize = (24, 10))
for i in range(10):
    ou_process = pd.Series(OrnsteinUhlenbeckProcess(5, 1).sample(1000, 1))
    ou_process.plot()
    
plt.grid(True)
plt.title("10 Ornstein Uhlenbeck Process with 5 speed revert")
plt.show()
```
